```python
import pandas as pd

# Load the iris dataset into a pandas DataFrame
data = pd.read_csv("C://Users\ketul\OneDrive\Documents\Iris.csv")

# Filter the DataFrame for each species
setosa_data = data[data['species'] == 'Iris-setosa']
versicolor_data = data[data['species'] == 'Iris-versicolor']
virginica_data = data[data['species'] == 'Iris-virginica']

# Display basic statistical details for each species
print("Species: Iris-setosa")
print("Mean:")
print(setosa_data.mean())
print("\nStandard Deviation:")
print(setosa_data.std())
print("\nPercentiles:")
print(setosa_data.quantile([0.25, 0.50, 0.75]))

print("\nSpecies: Iris-versicolor")
print("Mean:")
print(versicolor_data.mean())
print("\nStandard Deviation:")
print(versicolor_data.std())
print("\nPercentiles:")
print(versicolor_data.quantile([0.25, 0.50, 0.75]))

print("\nSpecies: Iris-virginica")
print("Mean:")
print(virginica_data.mean())
print("\nStandard Deviation:")
print(virginica_data.std())
print("\nPercentiles:")
print(virginica_data.quantile([0.25, 0.50, 0.75]))
```
