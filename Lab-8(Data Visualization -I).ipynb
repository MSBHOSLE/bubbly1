```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic
```

```python
sns.histplot(data=titanic, x='fare', kde=True)
plt.xlabel('Ticket Price')
plt.ylabel('Count')
plt.title('Distribution of Ticket Prices')
plt.show()
```

```python
import seaborn as sns

# Calculate the correlation matrix
correlation_matrix = titanic.corr()

# Plot the heatmap
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
import seaborn as sns

# Plot the count of passengers by a categorical variable, e.g., 'survived'
sns.countplot(data=titanic, x='survived')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.title('Passenger Survival')
plt.show()
```

```python
import seaborn as sns

# Plot a scatterplot of age against fare
sns.scatterplot(data=titanic, x='age', y='fare', hue='survived')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare (Colored by Survival)')
plt.show()
```
